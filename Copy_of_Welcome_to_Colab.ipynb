LISTS

```
num=[1,2,3,4,5]
num.append(6)            #append()-adds an item to the end if the list
print(num)
```

```
num=[1,2,3,4,5]
num.insert(3,8)           #insert()-adds an item in the given index value
print(num)
```

```
num=[1,2,3,4,5]
num.remove(5)             #remove()-removes the given number
print(num)
```

```
num=[1,2,3,4,5]
num.pop(4)               #pop()-removes the given index value. if index value not given then removes the last index value of the list
print(num)
```

```
num=[5,6,3,7,2,4]
num.sort()           #sort()- arranges the given list in assending order
print(num)
```

```
num=[5,6,3,7,2,4]
num.sort(reverse=True)         #sort(reverse=True)-arranges the list in desending order
print(num)
```

```
num=[5,6,3,7,2,4]
num.clear()        #clear()-clears the total list
print(num)
```

```
num=[5,6,3,7,2,4]
num.reverse()     #reverse()-reverses the list
print(num)
```

```
num=[5,6,3,7,2,4,5,6,4,6]
print(num.count(6))       #count()-prints the no:of times the given number is appaered
```

```
num=[5,6,3,7,2,4]
print(num.index(4))     #index()-prints the index value of the item
```

```
num=[5,6,3,7,2,4]
new_list=num.copy()      #copy()-copies the whole list
print(new_list)
```

```
num=[5,6,3,7,2,4]
del num[3:5]        #delete()-deletes the given index value from the list
print(num)
```

```
num=[5,6,3,7,2,4]
num.extend([2,4])      #extend([])-adds the multiple items to the list
print(num)
```

```
num=[5,6,3,7,2,4]
print(min(num))     #min()-prints the minimum value in the list
```

```
num=[5,6,3,7,2,4]
print(max(num))      #max()-prints the maximum value in the list
```

```
num=[5,6,3,7,2,4]
print(sum(num))      #sum()-prints the sum of the list
```
